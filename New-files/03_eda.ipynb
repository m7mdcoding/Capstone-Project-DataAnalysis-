```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def makehist(df,col,title=""): # A func to help make (making histograms) easier
  sns.histplot(data = df, x = df[col],kde=True)
  plt.title(title)
  plt.show()

df = pd.read_csv("/content/AmesHousingCleaned.csv")
```

```python
#TASK1
#1:
makehist(df,"Total Bsmt SF")
print("The data shows a right skewed distrubution")
#2
makehist(df,"Year Built")
print("The data shows a left skewed distrubution")
#3
makehist(df,"Mo Sold")
print("The data shows a symmetric skewed distrubution")
```

```python
#TASK2
#Category 1: Street affect on price
sns.boxplot(data = df, x= "Street", y = "SalePrice")
plt.show()
print("This chart shows us how the material of the street the house is on can affect its price")
#Category 2: zone affect on price
sns.boxplot(data = df, x= "MS Zoning", y = "SalePrice")
plt.show()
print("This chart shows us how the zone/area the house is in affects its price")
```

```python
#corrmatrix for TASK3
corr_matrix = df[[
"Year Built",
"Year Remod/Add",
"Mas Vnr Area",
"Total Bsmt SF",
"1st Flr SF",
"Gr Liv Area",
"Garage Cars",
"Garage Area",
"Full Bath",
"Overall Qual",
"SalePrice"]].corr()
```

```python
#TASK3
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax = 1)
plt.show()
print("This chart shows us how strong the correlation is between diffrent columns")
```

```python
#TASK4
plt.figure(figsize=(10,10))
sns.scatterplot( x = df["SalePrice"], y = df["Overall Qual"], hue = df["Year Built"]) #Price as X, Quality as y and the year as the colors
plt.show()
print("This chart shows us the price and quality of houses over time, we can see that the price and quality has gone up with time ")
```
