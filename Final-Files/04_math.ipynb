```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
std_scaler = StandardScaler()
df = pd.read_csv("/content/AmesHousingCleaned.csv")
df_std = df.copy()
```

```python
#TASK1
MEAN = np.mean(df["SalePrice"])
STD = np.std(df["SalePrice"])
```

```python
#TASK2
standrised_price = df_std["SalePrice_std"] = std_scaler.fit_transform(df[["SalePrice"]]) # The values after the standard scaler
bybroadcasting = (df["SalePrice"] - MEAN) / STD # The values after broadcasting

print(f"The value from the StandardScaler is\n{standrised_price}\n\nThe value from brodcasting is\n{bybroadcasting}") # comparing them
```

```python
features = df.select_dtypes(exclude=["object"]) # choosing only the numerical cols because you can't use cosine similarity on objects
```

```python
#TASK3
MAX = features.loc[df[["SalePrice"]].idxmax()] # idmax -> gives the id, loc -> gets the whole row of that id
MIN = features.loc[df[["SalePrice"]].idxmin()]
similarity = cosine_similarity(MAX, MIN) # Cosine similarity checks if the data is going in the same direction
print(similarity)
```

```python
#TASK4
NOVOT = 0
TNOV = (df["SalePrice"].count()) # Total Num Of Vals
for i in df["SalePrice"]:
  if i > 160000.000: # This threshold is the 50% of the col
    NOVOT += 1

print(f"The total num of values in the SalePrice col is: {TNOV}\nand the num of values over the threshold(which is 160k) is: {NOVOT}")
```
